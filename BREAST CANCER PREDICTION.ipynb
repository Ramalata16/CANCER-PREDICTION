```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
df
```

```python
df = df.drop(['id'], axis=1)
df
```

```python
df.isnull().sum()
```

```python
X = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size = 0.2, random_state=16)

X_train.shape
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3, min_samples_leaf=2, random_state = 16)
model.fit(X_train, y_train)
```

```python
ypred = model.predict(X_test)
ypred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix

accuracy = accuracy_score(ypred, y_test)
accuracy
```

```python
confusion_matrix(ypred, y_test)
```

```python
cbm = classification_report(ypred, y_test)
print(cbm)
```
